# Add Column by Example in Power Query

## Introduction

Column by Example in Power Query allows you to create new columns based on examples you provide, making it easier to generate complex transformations without deep knowledge of Power Query formulas.

## Import Data into Power Query

1. **Select the Dataset:**
   - Place your cursor within the dataset you want to work with.
   - Right-click and select `Get Data from Table/Range`.
   - Ensure the table has headers and the "marching ants" encompass the entire dataset.
   - Click `OK`.

## Adding a Column by Example

1. **Open the Power Query Editor:**
   - With your data loaded in Power Query, go to the `Add Column` tab.
   - Select `Column From Examples`.

2. **Provide Examples for the New Column:**
   - Start typing an example of what you want in the new column.
   - For instance, if you want a column with first names in uppercase and last name initials:
     - Type the desired result for the first row (e.g., `JAMES A.` for James Anderson).
   - Power Query will try to detect the pattern and apply it to the rest of the column.

3. **Refine the Example:**
   - If the initial example doesn't produce the correct result, provide more examples:
     - Continue typing examples for additional rows until Power Query correctly identifies the pattern.

4. **Review Generated Formula:**
   - Power Query will automatically generate a formula based on your examples.
   - You can see the formula in the formula bar, which can help you learn Power Query functions.

5. **Finalize and Apply the New Column:**
   - Once the pattern is correctly applied, click `OK`.
   - The new column will be added to your dataset.

## Examples of Columns by Example

1. **Create Name Badges:**
   - Example: Convert first names to uppercase and add the initial of the last name.
   - Type `JAMES A.` for James Anderson.
   - Provide additional examples if needed to refine the pattern.

2. **Combine Last Name and Extension:**
   - Example: Combine last name with an "x" followed by the extension, ensuring the extension is four digits.
   - Type `Arcaro x7304` for the first example.
   - For shorter extensions, provide examples with leading zeros (e.g., `Mendoza x0097`).

3. **Handle Null Values:**
   - Example: Create a column indicating preferences, and handle null values separately.
   - Type examples like `James likes Red-Gold` and `Nathan has no preference` for rows with null values.
   - If the pattern is not correctly identified, handle nulls separately.

4. **Extract Mentor Initials:**
   - Example: Extract initials from mentor names.
   - Type initials for the first few mentors (e.g., `MC` for Magnolia Charles).
   - Provide additional examples to help Power Query identify the correct pattern.

## Handling Issues and Refining Results

1. **Check the Generated Formula:**
   - Review the formula Power Query generates to understand how it derives the new column.
   - Make adjustments if necessary.

2. **Refine with Additional Examples:**
   - If Power Query struggles to identify the pattern, provide more examples or adjust existing ones.

3. **Manual Adjustments:**
   - For complex transformations, you may need to manually edit the generated formula or combine multiple steps.

## Final Steps

1. **Review and Adjust:**
   - Check the new column for accuracy and make any necessary adjustments.

2. **Close and Load Data:**
   - Once the data is prepared, click `Close & Load`.
   - Choose `Close & Load To` if you want to specify the load location:
     - **Existing Worksheet:** Select a specific cell in the current workbook.
     - **New Worksheet:** Load the data into a new worksheet.

By following these steps, you can effectively use Column by Example in Power Query to create new columns based on the patterns you provide, simplifying complex data transformations.
