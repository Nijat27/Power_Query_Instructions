# Sort and Filter Data in Power Query

## Introduction

Sorting and filtering data are essential operations in data analysis. Power Query offers robust features for sorting and filtering data, allowing for precise and customized data views. This guide provides general steps for sorting and filtering data in Power Query.

## Import Data into Power Query

1. **Open Excel and Load Data into Power Query:**
   - Launch Excel and open the workbook with the data.
   - Place your cursor within the dataset you want to work with.
   - Right-click and select `Get Data from Table/Range`.
   - Ensure the table has headers and the "marching ants" encompass the entire dataset.
   - Click `OK`.

## Sorting Data in Power Query

1. **Open the Power Query Editor:**
   - With your data loaded in Power Query, the Power Query Editor will open.

2. **Sort by Single Column:**
   - Select the column you want to sort by.
   - Go to the `Sort Ascending` or `Sort Descending` button in the toolbar to sort the column in the desired order.

3. **Sort by Multiple Columns:**
   - To sort by multiple levels, you must sort each column sequentially in the desired order:
     1. **First Column:** Select the first column (e.g., `Team`) and sort it (e.g., `Sort Ascending`).
     2. **Second Column:** Select the second column (e.g., `Year`) and sort it (e.g., `Sort Ascending`).
     3. **Third Column:** Select the third column (e.g., `Last Name`) and sort it (e.g., `Sort Ascending`).
     4. **Fourth Column:** Select the fourth column (e.g., `First Name`) and sort it (e.g., `Sort Ascending`).

4. **Result:**
   - Your data will now be sorted by the specified columns in the desired order.

## Filtering Data in Power Query

1. **Open the Power Query Editor:**
   - With your data loaded in Power Query, the Power Query Editor will open.

2. **Filter by Single Column:**
   - Click the filter drop-down button on the column you want to filter.
   - Select the filter criteria from the available options (e.g., `Equals`, `Does Not Equal`, `Contains`, `Begins With`, `Ends With`).

3. **Advanced Text Filters:**
   - For more complex filtering, select `Text Filters` from the drop-down menu.
   - Choose the desired filter option (e.g., `Equals`, `Does Not Equal`, `Contains`, `Begins With`, `Ends With`).
   - Enter the filter criteria and click `OK`.

4. **Filter by Multiple Columns:**
   - To filter multiple columns, apply filters sequentially to each column as needed:
     - **First Column:** Apply the desired filter (e.g., filter `Team` column to show only `Blue Herons` and `Goblins`).
     - **Second Column:** Apply another filter (e.g., filter `Year` column to show only values greater than or equal to 3).

5. **Using Advanced Filtering:**
   - For complex filtering logic, use the `Advanced Filter` option:
     - Select the column and click the filter drop-down.
     - Choose `Advanced Filter`.
     - Set up multiple conditions using `And`/`Or` logic (e.g., `Team equals Blue Herons OR Team equals Goblins` AND `Year is greater than or equal to 3`).

6. **Result:**
   - Your data will now be filtered according to the specified criteria.

## Final Steps

1. **Review and Adjust:**
   - Check the sorted and filtered data for accuracy and make any necessary adjustments.

2. **Close and Load Data:**
   - Once the data is prepared, click `Close & Load`.
   - Choose `Close & Load To` if you want to specify the load location:
     - **Existing Worksheet:** Select a specific cell in the current workbook.
     - **New Worksheet:** Load the data into a new worksheet.

By following these steps, you can effectively sort and filter your data in Power Query, allowing for precise and customized views of your data for analysis.
